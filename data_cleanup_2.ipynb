##Create sample dataset

```python
import pandas as pd
import numpy as np

# Original data
data = {
    'Name': ['John', 'Alice', 'Bob', 'Emily', 'David', 'Sarah', 'Michael', 'Eva', 'Daniel', 'Olivia', 'Bob', 'John', 'Eva'],
    'Age': [30, 25, 35, 28, 40, np.nan, 45, 27, 33, np.nan, 35, 30, 27],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Female', 'Male', 'Male', 'Female'],
    'Income': [50000, 60000, 70000, 55000, 65000, 55000, 75000, np.nan, 60000, 70000, 270000, 50000, np.nan],
    'City': ['New York', 'Los Angeles', 'Chicago', 'New York', 'Los Angeles', 'Chicago', 'New York', 'Los Angeles', 'Chicago', 'New York', 'Chicago', 'New York', 'Los Angeles']
}
```

##Load and read data to dataframe

```python

# Create DataFrame
df = pd.DataFrame(data)

# Display original dataset
print("Original Dataset:")
df
```

### Remove duplicates

```python
# Data Cleanup

df = df.drop_duplicates()
df
```

```python
# Handle missing values
df = df.dropna()
df
```

```python
# Mengimputasi missing value dengan mean
#mean_age = df['Age'].mean()
#df['Age'] = df['Age'].fillna(mean_age)
```

```python

# Handling data noise (outliers)
q1 = df['Income'].quantile(0.25)
q3 = df['Income'].quantile(0.75)
iqr = q3 - q1
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr
df = df[(df['Income'] >= lower_bound) & (df['Income'] <= upper_bound)]
df
```
